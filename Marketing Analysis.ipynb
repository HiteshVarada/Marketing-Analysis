```python
# prompt: import pandas, numpy and matplotlib

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# prompt: read the data Digital Marketing Data.csv

# Read the data
data = pd.read_csv('Digital Marketing Data.csv')
data.head()  # Display the first few rows to check
```

```python
# prompt: find the null values in the data dataframe

# Check for null values
data.isnull().sum()
```

```python
data.info()
```

```python
# prompt: describe the Data dataframe

data.describe()
```

```python
# prompt: Find the shape of the dataframe data

data.shape
```

```python
# prompt: Drop the columns 'Unnamed: 11' and 'Unnamed: 12' from the existing DataFrame data by not reading the data again

data = data.drop(['Unnamed: 11', 'Unnamed: 12'], axis=1)
```

```python
# prompt: show the dataframe data by not reading it

data
```

```python
# prompt: drop the null values from placement

data = data.dropna(subset=['placement'])
```

```python
# prompt: find the null values

data.isnull().sum()
```

```python
# prompt: value counts for campaign_name, banner, placement and also nunique values count

# Value counts for categorical variables
print("Campaign Name Value Counts:\n", data['campaign_name'].value_counts())
print("\nBanner Value Counts:\n", data['banner'].value_counts())
print("\nPlacement Value Counts:\n", data['placement'].value_counts())

# Unique value counts
print("\nUnique Campaign Names:", data['campaign_name'].nunique())
print("Unique Banners:", data['banner'].nunique())
print("Unique Placements:", data['placement'].nunique())
```

```python
# prompt: Compare the performance of different campaigns by looking at metrics such as revenue, cost, clicks, and post_click_conversions.

# Group by campaign name and aggregate relevant metrics
campaign_performance = data.groupby('campaign_name').agg({
    'revenue': 'sum',
    'cost': 'sum',
    'clicks': 'sum',
    'post_click_conversions': 'sum'
})

# Calculate key performance indicators (KPIs)
campaign_performance['CTR'] = (campaign_performance['clicks'] / campaign_performance['post_click_conversions']) * 100
campaign_performance['Conversion Rate'] = (campaign_performance['post_click_conversions'] / campaign_performance['clicks']) * 100
campaign_performance['Cost per Click (CPC)'] = campaign_performance['cost'] / campaign_performance['clicks']
campaign_performance['Return on Ad Spend (ROAS)'] = campaign_performance['revenue'] / campaign_performance['cost']

# Sort by revenue for better comparison
campaign_performance = campaign_performance.sort_values(by='revenue', ascending=False)

print(campaign_performance)

# Visualize the performance
plt.figure(figsize=(12, 6))
campaign_performance['revenue'].plot(kind='bar', color='skyblue')
plt.title('Revenue by Campaign')
plt.xlabel('Campaign Name')
plt.ylabel('Revenue')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# You can similarly visualize other metrics like cost, clicks, conversions, CTR, etc.
# For example:
# plt.figure(figsize=(12, 6))
# campaign_performance['CTR'].plot(kind='bar', color='green')
# plt.title('CTR by Campaign')
# ...
```

```python
# prompt: find how many clicks by banner, placement sort by clicks

# Group by banner and placement, and sum clicks
banner_placement_clicks = data.groupby(['banner', 'placement'])['clicks'].sum().reset_index()

# Sort by clicks in descending order
banner_placement_clicks = banner_placement_clicks.sort_values(by='clicks', ascending=False)

# Display the result
print(banner_placement_clicks)
```

```python
# prompt: campaign, user_engagement and count of user_engagement

# Group by campaign name and user engagement, and count the occurrences
campaign_user_engagement = data.groupby(['campaign_name', 'user_engagement'])['user_engagement'].count().reset_index(name='count')

# Display the result
print(campaign_user_engagement)
```

```python
# prompt: import datetime and convert data['day'] datatype to tmestamp

import datetime

# Assuming 'day' column is in a format that can be directly converted to datetime
data['day'] = pd.to_datetime(data['day'])

# If 'day' is in a different format, you might need to specify the format
# For example, if it's in the format 'YYYY-MM-DD':
# data['day'] = pd.to_datetime(data['day'], format='%Y-%m-%d')

data.info()  # Verify the datatype change
```

```python
# prompt: analyze how each campaign with day performed by no of clicks and post_click_conversion

# Group by campaign name, day, and aggregate clicks and post-click conversions
campaign_day_performance = data.groupby(['campaign_name', 'day']).agg({
    'clicks': 'sum',
    'post_click_conversions': 'sum'
}).reset_index()

# Display the result
print(campaign_day_performance)

# Visualize the performance of each campaign over time
import matplotlib.pyplot as plt

for campaign in data['campaign_name'].unique():
  campaign_data = campaign_day_performance[campaign_day_performance['campaign_name'] == campaign]
  plt.figure(figsize=(10, 6))
  plt.plot(campaign_data['day'], campaign_data['clicks'], label='Clicks')
  plt.plot(campaign_data['day'], campaign_data['post_click_conversions'], label='Conversions')
  plt.title(f'Clicks and Conversions for {campaign}')
  plt.xlabel('Day')
  plt.ylabel('Count')
  plt.legend()
  plt.xticks(rotation=45, ha='right')
  plt.tight_layout()
  plt.show()
```

```python
# prompt: Generate to see data['cost'] and data['post_click_sales_amount'] by day

# Group by day and aggregate cost and post-click sales amount
daily_cost_sales = data.groupby('day').agg({
    'cost': 'sum',
    'post_click_sales_amount': 'sum'
}).reset_index()

# Display the result
print(daily_cost_sales)

# Visualize the cost and sales over time
plt.figure(figsize=(12, 6))
plt.plot(daily_cost_sales['day'], daily_cost_sales['cost'], label='Cost')
plt.plot(daily_cost_sales['day'], daily_cost_sales['post_click_sales_amount'], label='Sales')
plt.title('Daily Cost and Sales')
plt.xlabel('Day')
plt.ylabel('Amount')
plt.legend()
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

By looking at the plotting we can see the amount spend for marketing has no change but the post_click_sales_amount are comin down.


```python
# prompt: generate a plot data['clicks'] by day

# Group by day and sum the clicks
daily_clicks = data.groupby('day')['clicks'].sum().reset_index()

# Plot the daily clicks
plt.figure(figsize=(12, 6))
plt.plot(daily_clicks['day'], daily_clicks['clicks'])
plt.title('Daily Clicks')
plt.xlabel('Day')
plt.ylabel('Number of Clicks')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

The clicks have also gone down

Checking if placement has to do anything with this.

```python
# prompt: generate a line plot with placement and no of clicks and post conversion clicks with day

# Group by placement, day, and sum clicks and post-click conversions
placement_day_performance = data.groupby(['placement', 'day']).agg({
    'clicks': 'sum',
    'post_click_conversions': 'sum'
}).reset_index()

# Plot for each placement
for placement in data['placement'].unique():
  placement_data = placement_day_performance[placement_day_performance['placement'] == placement]
  plt.figure(figsize=(10, 6))
  plt.plot(placement_data['day'], placement_data['clicks'], label='Clicks')
  plt.plot(placement_data['day'], placement_data['post_click_conversions'], label='Conversions')
  plt.title(f'Clicks and Conversions for {placement}')
  plt.xlabel('Day')
  plt.ylabel('Count')
  plt.legend()
  plt.xticks(rotation=45, ha='right')
  plt.tight_layout()
  plt.show()
```

The placement has no problem.

Lets find out if there is any relation between user engagement, banner and placement


```python
# prompt: Generate a clear line plot to find if there is any relation with user engagement,  banner
# . Import required packages.

import seaborn as sns

# Group by banner and user engagement, and count occurrences
banner_engagement = data.groupby(['banner', 'user_engagement'])['user_engagement'].count().reset_index(name='count')

# Plot using seaborn
plt.figure(figsize=(12, 6))
sns.lineplot(x='banner', y='count', hue='user_engagement', data=banner_engagement, markers=True, dashes=False)
plt.title('User Engagement by Banner')
plt.xlabel('Banner')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# prompt: Generate user engagement by day and no of clicks

# Group by day and user engagement, and sum clicks
day_engagement_clicks = data.groupby(['day', 'user_engagement'])['clicks'].sum().reset_index()

# Plot using seaborn
plt.figure(figsize=(12, 6))
sns.lineplot(x='day', y='clicks', hue='user_engagement', data=day_engagement_clicks, markers=True, dashes=False)
plt.title('User Engagement and Clicks by Day')
plt.xlabel('Day')
plt.ylabel('Clicks')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
